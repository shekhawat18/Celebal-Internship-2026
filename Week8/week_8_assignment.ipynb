# 🧠 Single Agent Pipeline Project

## Problem Statement
Build a **Single-Agent Smart Assistant** that:
- Understands user queries
- Routes tasks based on intent
- Uses tools when required
- Returns structured JSON output

### The agent should handle:
- Math queries → Calculator Tool
- Keyword extraction → Keyword Tool
- General queries → Direct response

---
### 🛠️ What You Need to Implement
- Agent logic
- Conditional routing
- Tool integration
- Basic error handling

### 🚀 Bonus
- Improve routing
- Add logging
- Add more tools


````python
# 🛠️ TOOL 1: Calculator

def calculator(expression: str) -> str:
    """Evaluate a mathematical expression."""
    try:
        return str(eval(expression))
    except Exception:
        return "Error in calculation"
````

````python
# 🛠️ TOOL 2: Keyword Extractor

def extract_keywords(text: str) -> list:
    """Extract keywords from text."""
    try:
        words = text.split()
        keywords = list(set([w.lower() for w in words if len(w) > 4]))
        return keywords[:5]
    except Exception:
        return []
````

````python
# 🛠️ TOOL 3: Word Counter

def word_counter(text: str) -> int:
    """Count the number of words in a sentence."""
    return len(text.split())
````

## 🤖 Implement Agent Logic Below

👉 Use conditional routing:
- If query contains "calculate" → use calculator
- If query contains "keywords" → use keyword extractor
- Else → general response

````python
#  BONUS TOOL: Word Counter
def word_counter(text: str) -> int:
    """Count the number of words in the given text."""
    return len(text.split())


# 🤖 AGENT FUNCTION
def agent(query: str):
    # Handle empty input
    if not query or not query.strip():
        return {
            "type": "error",
            "result": "Empty query received. Please enter a valid query."
        }

    query_lower = query.lower().strip()

    try:

        # =========================
        # Calculator Tool
        # =========================
        if (
            "calculate" in query_lower
            or any(op in query for op in ["+", "-", "*", "/"])
        ):

            print("[LOG] Calculator Tool Selected")

            expression = (
                query
                .replace("Calculate", "")
                .replace("calculate", "")
                .strip()
            )

            if not expression:
                return {
                    "type": "error",
                    "result": "No mathematical expression provided."
                }

            return {
                "type": "calculation",
                "result": calculator(expression)
            }

        # =========================
        # Keyword Extraction Tool
        # =========================
        elif (
            "keyword" in query_lower
            or "extract" in query_lower
        ):

            print("[LOG] Keyword Extraction Tool Selected")

            text = (
                query
                .replace("Extract keywords from", "")
                .replace("extract keywords from", "")
                .replace("Give keywords from", "")
                .replace("give keywords from", "")
                .strip()
            )

            if not text:
                return {
                    "type": "error",
                    "result": "No text provided for keyword extraction."
                }

            return {
                "type": "keywords",
                "result": extract_keywords(text)
            }

        # =========================
        # Word Counter Tool
        # =========================
        elif (
            "count words" in query_lower
            or "word count" in query_lower
        ):

            print("[LOG] Word Counter Tool Selected")

            text = (
                query
                .replace("Count words in", "")
                .replace("count words in", "")
                .replace("Word count", "")
                .replace("word count", "")
                .strip()
            )

            if not text:
                return {
                    "type": "error",
                    "result": "No text provided for word counting."
                }

            return {
                "type": "word_count",
                "result": word_counter(text)
            }

        # =========================
        # General Query
        # =========================
        else:

            print("[LOG] General Response Selected")

            return {
                "type": "general",
                "result": (
                    "Unsupported query. I currently support:\n"
                    "1. Calculator\n"
                    "2. Keyword Extraction\n"
                    "3. Word Counter"
                )
            }

    except Exception as e:

        print(f"[ERROR] {e}")

        return {
            "type": "error",
            "result": str(e)
        }
````

## 📦 Expected Output Format

```
{
  "type": "calculation / keywords / general / error",
  "result": ...
}
```

````python
queries = [
    # Calculator
    "Calculate 20 + 5",
    "20 * 15",
    "Calculate 10 / 0",      # Division by zero
    "Calculate",             # Empty expression

    # Keyword Extraction
    "Extract keywords from Artificial Intelligence is transforming industries",
    "Give keywords from Machine Learning is amazing",
    "Extract keywords from",  # Empty text

    # Word Counter
    "Count words in Artificial Intelligence is transforming industries",
    "Word count Artificial Intelligence is transforming industries",
    "Count words in",         # Empty text

    # General Queries
    "What is machine learning?",
    "Hello",

    # Invalid Inputs
    "",                       # Empty query
    "12345",
    "@#$%^&*"
]

for q in queries:
    print("=" * 60)
    print("Query:", repr(q))
    print("Response:", agent(q))
````

````python
# 🎯 Interactive Mode

print("=" * 60)
print(" Single Agent is Ready!")
print("Supported Commands:")
print("• Calculate 20 + 5")
print("• 20 * 15")
print("• Extract keywords from Artificial Intelligence is transforming industries")
print("• Count words in Artificial Intelligence is transforming industries")
print("Type 'exit' to stop.")
print("=" * 60)

while True:
    user_input = input("\nEnter query: ").strip()

    if user_input.lower() == "exit":
        print("\n Exiting Agent. Goodbye!")
        break

    response = agent(user_input)

    print("\nResponse:")
    print(response)
````
